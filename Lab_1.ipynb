```python
## Data Understanding - Verify data quality (Earl)

# Loading relevant libraries used throughout the analysis.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.tools.plotting import scatter_matrix

# Loading the "DelayedFlights.csv" data set
flights=pd.read_csv('C:\\Users\\Earl\\Documents\\MSDS 7331\\Data\\Lab 1\\airline_delays.csv')
# http://www.transtats.bts.gov/OT_Delay/OT_DelayCause1.asp?pn=1

# Number of rows in data set (54,789 rows total)
```

```python
flights.head(10) # Initial view of data set
# year: The year in which the data was collected.
# month: The month in which the data was collected.
# carrier: The airline code, i.e. two-character IATA code
# carrier_name: The airline name
# airport: The airport code
# airport_name: The airport name
# arr_flights: # of flights that arrived at the airport
# arr_del15: # of flights that arrived>=15 minutes late
# carrier_ct: # of flights delayed due to the carrier
# weather_ct: # of flights delayed due to the weather
# nas_ct: # of flights delayed due to the national aviation system (e.g. non-extreme weather conditions, 
# airport operations, heavy traffic volume, and air traffic control)
# security_ct: # of flights delayed due to security
# late_aircraft_ct: # of flights delayed because a previous flight using the same aircraft was late
# arr_cancelled: # of canceled arrivals
# arr_diverted: # of scheduled arrivals that were diverted
# arr_delay: sum of the delay in minutes
# carrier_delay: Total minutes of delays due to carriers
# weather_delay: Total minutes of delays due to weather
# nas_delay: Total minutes of delays due to Natl. Air Service
# security_delay: Total minutes of delays due to security
# late_aircraft_delay: Total minues of delay due to a previous flight using the same aircraft arriving late
# Source: http://www.cs.siue.edu/~marmcke/docs/cybergis/mapReduceAirline.html


```

```python
# Checking variable types of features in the data set and examining counts
print ('----------------------------------------------------')
print ('Data Types and Missing Values Prior to Data Handling')
print ('----------------------------------------------------')
print (flights.info()) # Summary of data set

# Missing Value Handling
# ----------------------
# 15 of the 21 features (~71%) in the "Airline Delay" data have missing observations. However, the percentage of
# missing values, relative to total observation count is minimal. For example, the feature with the most 
# missing values, "arr_del15" has roughly 0.14% of its observations missing. Considering "arr_del15" is likely a
# primary candidate as the dependent/response variable in future classification and/or predictive efforts and 
# the frequency of missing values is rather minimal, the best course of action is to remove them.

# Removing observations with missing values and re-checking total observation count
flights2=flights.dropna() # Removing observations with missing values
print ('')
print ('-------------------------------------------------')
print ('Data Types and Missing Values After Data Handling')
print ('-------------------------------------------------')
print (flights2.info()) # Summary of data set after removing missing values
print ('**Note: 54,711 (~99.9%) observations remain.')

# After removing missing values, 54,711 (~99.9%) observations remain.
```

```python
# Summary Statistics of Continous Vars of Data after Removing Missing Observations
print ('--------------------------------------')
print ('Summary Statistics for Continuous Vars')
print ('--------------------------------------')
(flights2.describe())
```

```python
# Checking median values before and after outlier removal to ensure data didin't change substantially.
print ('----------------------------------------------')
print ('Median Values Before Removal of Missing Values')
print ('----------------------------------------------')
print (flights.median())
print ('')
print ('---------------------------------------------')
print ('Median Values After Removal of Missing Values')
print ('---------------------------------------------')
print (flights2.median())
print ('**Note: No substantial changes in data integrity.')
```

```python
# Summary Statistics - Avg Flight Arrivals by Year
# i.e. How many flights arrive in a given year, on average?
flights_grouped_by_year=flights2.groupby(by='year') # Creating group by 'year'
print ('')
print ('----------------------------------------------------')
print ('Avg Flight Arrivals by Year')
print ('----------------------------------------------------')
print ('*i.e. How many flights arrived per year, on average?')
print (flights_grouped_by_year.arr_flights.mean())  # Avg of flight arrivals by 'year', overall

# Summary Statistics - Avg Delayed Flights by Year
# i.e. How many flights were delayed in a given year, on average?
flights_grouped_by_year=flights2.groupby(by='year') # Creating group by 'year'
print ('')
print ('---------------------------------------------------------')
print ('Avg Delayed Flights by Year')
print ('---------------------------------------------------------')
print ('*i.e. How many flights were delayed per year, on average?')
print (flights_grouped_by_year.arr_del15.mean())  # Avg of delayed flights by 'year', overall

# Summary Statistics - % of Delayed Flights by Year
# i.e. What percentage of flights in a given year were delayed
flights_grouped_by_year=flights.groupby(by='year') # Creating group by 'year'
print ('')
print ('---------------------------------------------------------------')
print ('% of Delayed Flights by Year')
print ('---------------------------------------------------------------')
print ('*i.e. What percentage of flights from a given year were delayed?')
print (flights_grouped_by_year.arr_del15.sum()/flights_grouped_by_year.arr_flights.sum())*100.0  # % of delayed flights by 'year', overall

# Summary Statistics - % of Diverted Flights by Year
# i.e. What percentage of flights in a given year were diverted
#flights_grouped_by_year=flights.groupby(by='year') # Creating group by 'year'
#print ('')
#print ('----------------------------------------------------------------')
#print ('% of Diverted Flights by Year')
#print ('----------------------------------------------------------------')
#print ('*i.e. What percentage of flights from a given year were diverted?')
#print(flights_grouped_by_year.arr_diverted.sum()/flights_grouped_by_year.arr_flights.sum())*100.0  # % of delayed flights by 'year', overall

```
